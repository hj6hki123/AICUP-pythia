```python
import re
import pandas as pd

train_phi_category = ['PATIENT', 'DOCTOR', 'USERNAME',
             'PROFESSION', 'ROOM', 'DEPARTMENT', 'HOSPITAL', 'ORGANIZATION', 'STREET', 'CITY', 'STATE', 'COUNTRY', 'ZIP', 'LOCATION-OTHER',
             'AGE', 'DATE', 'TIME', 'DURATION', 'SET',
             'PHONE', 'FAX', 'EMAIL', 'URL', 'IPADDR',
             'SSN', 'MEDICALRECORD', 'HEALTHPLAN', 'ACCOUNT', 'LICENSE', 'VEHICLE', 'DEVICE', 'BIOID', 'IDNUM','PHI']

with open("llama3/generated_sentences.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()
    
    context = []
    category = []
    word = []
    
    for line in lines:
        if "\t" in line:
            left, right = line.split("\t", 1    )
            if ":" in right:
                cleft, cright = right.split(":", 1)
                context.append(left.strip())
                category.append(cleft.strip())
                word.append(cright.strip())


    
    df = pd.DataFrame({'Context': context, 'Category': category, 'Word': word})
    
    df = df.drop_duplicates()
    
    df = df[df['Category'].isin(train_phi_category)]
    df.reset_index(drop=True, inplace=True)
    
df
```

```python
with open("llama3/modified_sentences.txt", "w", encoding="utf-8") as f:
    for index, row in df.iterrows():
        f.write(f"{row['Context']}\t{row['Category']}: {row['Word']}\n")
```
